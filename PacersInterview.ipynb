```python
olsDFprep = pd.merge(teamClutchAdvanced, teamGeneral, how='inner', left_index=True, right_index=True, on='TEAM')
olsDFprep = olsDFprep.dropna(axis=1, how='all')
olsDF = olsDFprep.filter(['TEAM','WIN%','NetRtg', 'eFG%'], axis = 1)
olsDF.head()
```

```python
x = olsDF['NetRtg'].values
y = olsDF['WIN%'].values
regressor = linear_model.LinearRegression()
regressor.fit(x.reshape(-1,1), y)
y_pred = regressor.predict(x.reshape(-1,1))
plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x.reshape(-1,1)), color = 'blue')
plt.title('NetRtg vs WIN%')
plt.xlabel('NetRtg')
plt.ylabel('WIN%')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(olsDF[['NetRtg', 'eFG%']], olsDF['WIN%'])
print(reg.coef_)
```
